指定求和汇总张量的轴

```python
A_sum_axis0 = A.sum(axis=0) #在第一个维度进行求和
A,A_sum_axis0, A_sum_axis0.shape
```

```python
A_sum_axis1 = A.sum(axis=1) #在第二个维度进行求和
A_sum_axis1, A_sum_axis1.shape
```

```python
A.sum(axis=[0,1]) #在第一个和第二个维度进行求和
```

```python
A.mean(dtype=torch.float32), A.sum()/A.numel()
```

按某个维度求和时不把它丢掉(可以用于广播机制，保持维度一致)

```python
sum_A = A.sum(axis=1,keepdim=True) #在第二个维度进行求和，并保持维度不变
sum_A, sum_A.shape
```

```python
A/sum_A
```

某个轴计算A元素的累计总和

```python
A=torch.arange(20).reshape(5,4)
A,A.cumsum(axis=0) #在第一个维度上累积求和
```

矩阵向量积

```python
x = torch.arange(4)
A.shape, x.shape, torch.mv(A,x) #矩阵-向量积
```

矩阵乘法

```python
B = torch.ones(4,3)
torch.mm(A.float(),B) 
```
